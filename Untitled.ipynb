```python
#Some numpy code
r = 0.025 #interest rate
T = 50 #end date
b = np.empty(T+1) #numpy array, for storing balance in each period.
b[0] = 10 #initial balance

for t in range(T):
    #note ability to iterate b[0]. [0] is like filter in numpy array (works same way in pandas?).
    b[t+1] = (1+r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```
